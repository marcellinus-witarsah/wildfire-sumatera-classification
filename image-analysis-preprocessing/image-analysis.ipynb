```python
def plot_bands(list_band_names, satellite_image_bands):
    fig, ax = plt.subplots(nrows=len(list_band_names), figsize=(10,30))

    for i in range(len(list_band_names)):
        sns.histplot(satellite_image_bands[i].reshape(-1), bins=256, ax=ax[i])
        ax[i].set_title(list_band_names[i])
    
    plt.show()
```

```python
landsat_postfire_band_names = ['DELTA_NBR', 'B7', 'B5', 'B4', 'B3', 'B2']
plot_bands(landsat_postfire_band_names, landsat_postfire_bands)
```
